# Model - what is our baseline ?
What is the simplest model we can think of ! 
In this case, if our baseline model is just randomly guessing the species of flower, or guessing a single specie for every data point, we would expect to have a model accuracy of 0.33 or 33%, since we have 3 different classes that are evenly balanced

So our model should at least beat 33% accuracy 

# Modeling - simple manual model 
Lets manually look at our data and decide some cutoff points for classification

```python
def single_feature_pred(petal_lenght):
    """Predict the species based on petal lenght"""
    if petal_lenght < 2.5:
        return 0
    elif petal_lenght < 4.8:
        return 1    
    else:
        return 2
```

```python
manual_model_accuracy = np.mean(manual_y_pred == y_train)
manual_model_accuracy
```

```python
def plot_incorrect_predictions (df_prediction, x_axis_featuere, y_axis_feature):
    fig, axs = plt.subplots(2, 2, figsize=(10,10))
    axs = axs.flatten()
    sns.scatterplot(x=x_axis_featuere, y=y_axis_feature, hue='prediction_label', data=df_prediction, ax=axs[0])
    sns.scatterplot(x=x_axis_featuere, y=y_axis_feature, hue='target_name', data=df_prediction, ax=axs[1])
    sns.scatterplot(x=x_axis_featuere, y=y_axis_feature, hue='Correct_prediction', data=df_prediction, ax=axs[2])
    axs[3].set_visible(False)
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10,10))
```

```python
plot_incorrect_predictions(df_prediction, 'petal length (cm)', 'petal width (cm)')
plt.show()
```

# Model Tunning 
What is model tunning
Model tuning is trying the parameters of your mdoel (these are 'hyperparameters') that maximize the model performance

```python
for reg_param in [10 ]:
    model = LogisticRegression(max_iter=250, C=reg_param)
    accuracies = cross_val_score(model, x_train, y_train, cv=5, scoring='accuracy')
    print(f"manual_model_accuracy: {manual_model_accuracy:.2%}")
    print(f"Cross val accuracy: {np.mean(accuracies):.2%}")
```

```python
# Import necessary library
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Initialize and train the model
dt_model = DecisionTreeClassifier()
dt_model.fit(x_train, y_train)

# Predict on test set
y_test_pred_dt = dt_model.predict(x_test)

# Calculate accuracy
dt_accuracy = accuracy_score(y_test, y_test_pred_dt)
print(f"Decision Tree Accuracy: {dt_accuracy:.2%}")
```

```python
# Import necessary library
from sklearn.ensemble import RandomForestClassifier

# Initialize and train the model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(x_train, y_train)

# Predict on test set
y_test_pred_rf = rf_model.predict(x_test)

# Calculate accuracy
rf_accuracy = accuracy_score(y_test, y_test_pred_rf)
print(f"Random Forest Accuracy: {rf_accuracy:.2%}")
```

```python
# Import necessary library
from sklearn.neighbors import KNeighborsClassifier

# Initialize and train the model
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(x_train, y_train)

# Predict on test set
y_test_pred_knn = knn_model.predict(x_test)

# Calculate accuracy
knn_accuracy = accuracy_score(y_test, y_test_pred_knn)
print(f"KNN Accuracy: {knn_accuracy:.2%}")
```

```python
# Import necessary library
from sklearn.svm import SVC

# Initialize and train the model
svm_model = SVC(kernel='linear')
svm_model.fit(x_train, y_train)

# Predict on test set
y_test_pred_svm = svm_model.predict(x_test)

# Calculate accuracy
svm_accuracy = accuracy_score(y_test, y_test_pred_svm)
print(f"SVM Accuracy: {svm_accuracy:.2%}")
```

```python
# Import necessary library
from xgboost import XGBClassifier

# Initialize and train the model
xgb_model = XGBClassifier(use_label_encoder=False, eval_metric='mlogloss')
xgb_model.fit(x_train, y_train)

# Predict on test set
y_test_pred_xgb = xgb_model.predict(x_test)

# Calculate accuracy
xgb_accuracy = accuracy_score(y_test, y_test_pred_xgb)
print(f"XGBoost Accuracy: {xgb_accuracy:.2%}")
```
